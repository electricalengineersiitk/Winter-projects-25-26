```python
import torch
import numpy as np
# ===========================================================
# PART 1: Data Generation (The Exclusion Zone)
# ===========================================================

# ⚠️ INSTRUCTION: Replace 1 with your actual Group Number
GROUP_NUMBER = 8

# Set seed for reproducibility
torch.manual_seed(GROUP_NUMBER)
np.random.seed(GROUP_NUMBER)

def generate_data(n_samples=3000):
    """
    Generates 3000 sensor readings for the Kepler-186f Exclusion Zone.
    Shape: Concentric circles (Donut).
    Logic:
      - Dead Zone (1): 2km < distance < 4km
      - Safe Zone (0): distance < 2km OR distance > 4km
    """
    # Generate random coordinates between -5 and 5 km
    X = (torch.rand(n_samples, 2) * 10) - 5

    # Calculate distance from center (radius)
    radius = torch.sqrt(X[:, 0]**2 + X[:, 1]**2)

    # Assign Labels: 1 if inside the Dead Zone, 0 otherwise
    y = ((radius > 2) & (radius < 4)).float().view(-1, 1)

    # Add 5% Noise (Malfunctioning Drones)
    n_noise = int(0.05 * n_samples)
    noise_indices = torch.randperm(n_samples)[:n_noise]
    y[noise_indices] = 1 - y[noise_indices] # Flip labels

    return X, y

# Generate the dataset
X_full, y_full = generate_data(3000)

print(f"Data Generated: {X_full.shape} samples.")
print(f"Target Generated: {y_full.shape} labels.")
```

```python
# Training and validation of sample

import matplotlib.pyplot as plt

n_total = X_full.shape[0]
cut = int(0.7 * n_total)

X_train = X_full[:cut]
y_train = y_full[:cut]

X_val = X_full[cut:]
y_val = y_full[cut:]

print("Train samples:", X_train.shape[0], "Val samples:", X_val.shape[0])
```

```python
# Initialization

# architecture 2 -> 16 -> 16 -> 16 -> 1
w1 = torch.randn(2, 16) * 0.01
b1 = torch.zeros(1, 16)
w2 = torch.randn(16, 16) * 0.01
b2 = torch.zeros(1, 16)
w3 = torch.randn(16, 16) * 0.01
b3 = torch.zeros(1, 16)
w4 = torch.randn(16, 1) * 0.01
b4 = torch.zeros(1, 1)


w1.requires_grad_()
b1.requires_grad_()
w2.requires_grad_()
b2.requires_grad_()
w3.requires_grad_()
b3.requires_grad_()
w4.requires_grad_()
b4.requires_grad_()

print("Weights and biases created.")
```

```python
# Activation

def relu(x):
    return x * (x > 0).float()

def sigmoid(x):
    return 1.0 / (1.0 + torch.exp(-x))
```

```python
# binary cross-entropy

def bce(y_true, y_pred):

    loss = - (y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred))
    return loss.mean()
```

```python
# Forward function

def net(X):
    z1 = X @ w1 + b1    # (N,16)
    a1 = relu(z1)
    z2 = a1 @ w2 + b2
    a2 = relu(z2)
    z3 = a2 @ w3 + b3
    a3 = relu(z3)
    z4 = a3 @ w4 + b4
    out = sigmoid(z4)   # (N,1)
    return out
```

```python
# Accuracy function

def accuracy(X, y):
    yhat = net(X)
    preds = (yhat > 0.5).float()
    return (preds == y).float().mean().item()

print("Final Train accuracy:", accuracy(X_train, y_train))
print("Final Validation accuracy:  ", accuracy(X_val, y_val))
```

```python
# Training

lr = 0.01
epochs = 1500

train_losses = []
val_losses = []
train_accs = []
val_accs = []

for epoch in range(epochs):
    y_hat = net(X_train)
    loss = bce(y_train, y_hat)

    loss.backward()

    params = [w1, b1, w2, b2, w3, b3, w4, b4]
    for p in params:
        p.data -= lr * p.grad.data
        p.grad.data.zero_()

    with torch.no_grad():
        t_preds = (net(X_train) > 0.5).float()
        t_acc = (t_preds == y_train).float().mean().item()
        train_accs.append(t_acc)

        yv = net(X_val)
        vloss = bce(y_val, yv)
        v_preds = (yv > 0.5).float()
        v_acc = (v_preds == y_val).float().mean().item()

        train_losses.append(loss.item())
        val_losses.append(vloss.item())
        val_accs.append(v_acc)

    if epoch % 500 == 0 or epoch == epochs - 1:
        print(f"Epoch {epoch:4d} | train_loss={loss.item():.4f} | validation_loss={vloss.item():.4f} | val_acc={v_acc:.4f}")

print("Final Train accuracy:", train_accs[-1])
print("Final Validation accuracy:  ", val_accs[-1])
```

```python
# Loss Plot
plt.figure(figsize=(6,4))
plt.plot(train_losses, label='Train')
plt.plot(val_losses, label='Validation')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('BCE Loss')
plt.legend()
plt.show()
```

```python
# Accuracy Plot

plt.figure(figsize=(6,4))
plt.plot(train_accs, label='Train')
plt.plot(val_accs, label='Validation')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Decision Boundary

xx, yy = np.meshgrid(np.linspace(-6, 6, 300), np.linspace(-6, 6, 300))
grid = np.c_[xx.ravel(), yy.ravel()]
grid_tensor = torch.tensor(grid, dtype=torch.float32)

zz = net(grid_tensor).detach().numpy().reshape(xx.shape)

plt.figure(figsize=(6,6))
plt.contourf(xx, yy, zz, levels=[0, 0.5, 1], alpha=0.6, cmap='coolwarm')
plt.scatter(X_val[:, 0], X_val[:, 1], c=y_val.squeeze(), s=5)
plt.title("Decision Boundary (Validation Set)")
plt.show()
```
