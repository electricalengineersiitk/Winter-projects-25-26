```python
import torch
import numpy as np
# ===========================================================
# PART 1: Data Generation (The Exclusion Zone)
# ===========================================================

# ⚠️ INSTRUCTION: Replace 1 with your actual Group Number
GROUP_NUMBER = 1

# Set seed for reproducibility
torch.manual_seed(GROUP_NUMBER)
np.random.seed(GROUP_NUMBER)

def generate_data(n_samples=3000):
    """
    Generates 3000 sensor readings for the Kepler-186f Exclusion Zone.
    Shape: Concentric circles (Donut).
    Logic:
      - Dead Zone (1): 2km < distance < 4km
      - Safe Zone (0): distance < 2km OR distance > 4km
    """
    # Generate random coordinates between -5 and 5 km
    X = (torch.rand(n_samples, 2) * 10) - 5

    # Calculate distance from center (radius)
    radius = torch.sqrt(X[:, 0]**2 + X[:, 1]**2)

    # Assign Labels: 1 if inside the Dead Zone, 0 otherwise
    y = ((radius > 2) & (radius < 4)).float().view(-1, 1)

    # Add 5% Noise (Malfunctioning Drones)
    n_noise = int(0.05 * n_samples)
    noise_indices = torch.randperm(n_samples)[:n_noise]
    y[noise_indices] = 1 - y[noise_indices] # Flip labels

    return X, y

# Generate the dataset
X_full, y_full = generate_data(3000)

print(f"Data Generated: {X_full.shape} samples.")
print(f"Target Generated: {y_full.shape} labels.")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Train / Validation Split

X_train, X_val, y_train, y_val = train_test_split(
    X_full.numpy(), y_full.numpy(), test_size=0.3, random_state=GROUP_NUMBER
)

X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_val   = torch.tensor(X_val, dtype=torch.float32)
y_val   = torch.tensor(y_val, dtype=torch.float32)
```

```python
# MLP from Scratch (3 Hidden Layers)

def init_weights(in_dim, out_dim):
    w = torch.randn(in_dim, out_dim) * 0.1
    b = torch.zeros(1, out_dim)
    w.requires_grad_(True)
    b.requires_grad_(True)
    return w, b

# Architecture: 2 → 16 → 16 → 16 → 1
W1, b1 = init_weights(2, 16)
W2, b2 = init_weights(16, 16)
W3, b3 = init_weights(16, 16)
W4, b4 = init_weights(16, 1)

params = [W1, b1, W2, b2, W3, b3, W4, b4]

# Activation Functions
def relu(z):
    return torch.clamp(z, min=0)

def sigmoid(z):
    return 1 / (1 + torch.exp(-z))

# Forward Pass
def forward(X):
    z1 = torch.matmul(X, W1) + b1
    a1 = relu(z1)

    z2 = torch.matmul(a1, W2) + b2
    a2 = relu(z2)

    z3 = torch.matmul(a2, W3) + b3
    a3 = relu(z3)

    z4 = torch.matmul(a3, W4) + b4
    y_hat = sigmoid(z4)

    return y_hat
```

```python
# Binary Cross Entropy

def binary_cross_entropy(y_hat, y):
    eps = 1e-7
    y_hat = torch.clamp(y_hat, eps, 1 - eps)
    loss = -torch.mean(y * torch.log(y_hat) + (1 - y) * torch.log(1 - y_hat))
    return loss

# Accuracy
def accuracy(y_hat, y):
    preds = (y_hat >= 0.5).float()
    return (preds == y).float().mean()
```

```python
# Training Loop

epochs = 300
lr = 0.01

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(epochs):

    # TRAIN
    y_hat_train = forward(X_train)
    train_loss = binary_cross_entropy(y_hat_train, y_train)
    train_acc = accuracy(y_hat_train, y_train)

    train_loss.backward()

    # SGD Update
    with torch.no_grad():
        for p in params:
            p -= lr * p.grad
            p.grad.zero_()

    # VALIDATION
    with torch.no_grad():
        y_hat_val = forward(X_val)
        val_loss = binary_cross_entropy(y_hat_val, y_val)
        val_acc = accuracy(y_hat_val, y_val)

    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())
    train_accs.append(train_acc.item())
    val_accs.append(val_acc.item())

    if epoch % 50 == 0:
        print(f"Epoch {epoch:03d} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
```

```python
# GDC HUD — 3 Panel Visualization

# System Error Trajectory (Loss Plot)
plt.figure(figsize=(7, 5))
plt.plot(train_losses, label="Training Loss", linewidth=2)
plt.plot(val_losses, label="Validation Loss", linewidth=2)
plt.xlabel("Epochs")
plt.ylabel("Binary Cross Entropy Loss")
plt.title("System Error Trajectory")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Integrity Check (Accuracy Plot)

plt.figure(figsize=(7, 5))
plt.plot(train_accs, label="Training Accuracy", linewidth=2)
plt.plot(val_accs, label="Validation Accuracy", linewidth=2)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Integrity Check: Train vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Geospatial Threat Map (Decision Boundary)

plt.figure(figsize=(7, 7))

xx, yy = np.meshgrid(
    np.linspace(-5, 5, 300),
    np.linspace(-5, 5, 300)
)

grid = torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32)

with torch.no_grad():
    probs = forward(grid).reshape(xx.shape)

plt.contourf(xx, yy, probs, levels=[0, 0.5, 1], alpha=0.6)
plt.scatter(
    X_val[:, 0],
    X_val[:, 1],
    c=y_val.squeeze(),
    s=12,
    edgecolors='k'
)

plt.xlabel("X (km)")
plt.ylabel("Y (km)")
plt.title("Geospatial Threat Map (Decision Boundary)")
plt.gca().set_aspect("equal")
plt.show()
```
