To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# @title
import cv2
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np
def convolve(img,kernel):

  img = img.astype(float)


  kernel = np.flipud(np.fliplr(kernel))
  #getting dimensions of kernel and image
  height_img,width_img=img.shape
  height_k,width_k=kernel.shape
  #height and width for padding
  height_pad=height_k//2
  width_pad=width_k//2

  padded_img = np.pad(img, ((height_pad, height_pad), (width_pad,width_pad)), mode='reflect')

  cout=np.zeros_like(img)

  for i in range(height_img):
        for j in range(width_img):

            R = padded_img[i : i + height_k, j : j + width_k]

            cout[i, j] = (R * kernel).sum()

  return cout
````

````python
def R_map(img, window_size=3, k=0.04):
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  Gx,Gy=gx_gy_calc(gray)

  pad=window_size//2

  h,w=gray.shape

  gxx=Gx**2
  gyy=Gy**2
  gxy=Gx*Gy

  R=np.zeros_like(gray, dtype=np.float32)
  for i in range(pad, h-pad):
    for j in range(pad, w-pad):
      sum1 = np.sum(gxx[i-pad:i+pad+1 , j-pad:j+pad+1])
      sum2 = np.sum(gyy[i-pad:i+pad+1 , j-pad:j+pad+1])
      sum3 = np.sum(gxy[i-pad:i+pad+1 , j-pad:j+pad+1])

      det = sum1*sum2 - (sum3**2)
      trace = sum1 + sum2

      R[i,j] = det - k*(trace**2)



  return R

def plot_R_map(img, name, window_size, k):
  R = R_map(img, window_size, k)

  plt.figure()
  plt.imshow(R, cmap='gray')
  plt.title(f"R map of {name}")
  plt.axis("off")

  images = {
    "flat image": flat,
    "edge image": edge,
    "corner image": corner,

  }
window_size = 3
k = 0.04

for key, value in images.items():
  plot_R_map(value, key,window_size,k)
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def hough_transform_mc(img, m_min=-5, m_max=5, m_bins=200, c_bins=200):
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  edges = cv2.Canny(gray, 100, 200)

  h, w = gray.shape

  m_vals = np.linspace(m_min, m_max, m_bins)
  c_vals = np.linspace(-h, h, c_bins)

  accumulator = np.zeros((m_bins, c_bins), dtype=np.float32)
  ys, xs = np.nonzero(edges)

  # voting in mc space
  for y, x in zip(ys, xs):
    for i, m in enumerate(m_vals):
      c = y - m * x
      j = int(np.round((c + h) / (2 * h) * (c_bins - 1)))
      if 0 <= j < c_bins:
        accumulator[i, j] += 1

  return accumulator, m_vals, c_vals


def get_lines_mc(acc, m_vals, c_vals, thresh_norm=0.7):
  acc_norm = acc / acc.max()
  indices = np.where(acc_norm > thresh_norm)

  lines = []
  for i, j in zip(indices[0], indices[1]):
    lines.append((m_vals[i], c_vals[j]))

  return lines


def draw_lines_mc(img, lines):
  # draw detected lines on image
  out = img.copy()
  h, w = img.shape[:2]

  for m, c in lines:
    x1, x2 = 0, w
    y1 = int(m * x1 + c)
    y2 = int(m * x2 + c)
    cv2.line(out, (x1, y1), (x2, y2), (255, 0, 0), 3)

  return out


corner2 = np.asarray(img_np, dtype=np.uint8)

acc, m_vals, c_vals = hough_transform_mc(corner2)
lines = get_lines_mc(acc, m_vals, c_vals, thresh_norm=0.79)
result = draw_lines_mc(corner2, lines)


plt.imshow(result)
plt.title("Detected Edges")
plt.axis("off")
plt.show()
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def hough_transform_rhotheta(img, theta_bins=180):
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  edges = cv2.Canny(gray, 100, 200)

  h, w = gray.shape
  diag_len = int(np.sqrt(h * h + w * w))

  thetas = np.linspace(0, np.pi, theta_bins)
  rhos = np.linspace(-diag_len, diag_len, 2 * diag_len)

  accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

  ys, xs = np.nonzero(edges)

  for y, x in zip(ys, xs):
    for t_idx, theta in enumerate(thetas):
      rho = x * np.cos(theta) + y * np.sin(theta)
      r_idx = int(np.round(rho + diag_len))
      if 0 <= r_idx < len(rhos):
        accumulator[r_idx, t_idx] += 1

  return accumulator, rhos, thetas


def get_lines_rhotheta(acc, rhos, thetas, thresh_norm=0.4):
  acc_norm = acc / acc.max()
  indices = np.where(acc_norm > thresh_norm)

  lines = []
  for r_idx, t_idx in zip(indices[0], indices[1]):
    lines.append((rhos[r_idx], thetas[t_idx]))

  return lines


def draw_lines_rhotheta(img, lines):
  out = img.copy()
  h, w = img.shape[:2]

  for rho, theta in lines:
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho

    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))

    cv2.line(out, (x1, y1), (x2, y2), (255, 0, 0), 2)

  return out


img = np.asarray(img_np, dtype=np.uint8)

acc_rt, rhos, thetas = hough_transform_rhotheta(img)
lines_rt = get_lines_rhotheta(acc_rt, rhos, thetas, thresh_norm=0.38)
result_rt = draw_lines_rhotheta(img, lines_rt)

plt.imshow(result_rt)
plt.title("Detected Lines ")
plt.axis("off")
plt.show()
````
