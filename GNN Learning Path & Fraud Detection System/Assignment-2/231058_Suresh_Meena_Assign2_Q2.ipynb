```python
import numpy as np

class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        # Convert y values to -1 and +1
        y = np.where(y <= 0, -1, 1)

        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0

        # Gradient Descent
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):

                # Margin condition: y_i (wÂ·x_i - b) >= 1 ?
                condition = y[idx] * (np.dot(x_i, self.w) - self.b) >= 1

                if condition:
                    # Only regularization
                    dw = 2 * self.lambda_param * self.w
                    db = 0
                else:
                    # Regularization + hinge loss gradient
                    dw = 2 * self.lambda_param * self.w - y[idx] * x_i
                    db = y[idx]

                # Update parameters
                self.w -= self.lr * dw
                self.b -= self.lr * db

    def predict(self, X):
        # Linear output
        approx = np.dot(X, self.w) - self.b

        # Return class labels (-1 or +1)
        return np.sign(approx)





#optional
if __name__ == "__main__":
    X = np.array([[1, 2], [2, 3], [3, 3], [2, 1]])
    y = np.array([1, 1, -1, -1])

    svm = LinearSVM(learning_rate=0.001, lambda_param=0.01, n_iters=1000)
    svm.fit(X, y)

    print("Weights:", svm.w)
    print("Bias:", svm.b)
    print("Predictions:", svm.predict(X))
```
