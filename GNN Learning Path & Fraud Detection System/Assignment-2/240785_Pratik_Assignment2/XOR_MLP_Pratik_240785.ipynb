```python
import numpy as np

x = np.array([[0, 0, 1, 1], [0, 1, 0, 1]])
y_true = np.array([[0, 1, 1, 0]])

np.random.seed(42)
W_1 = np.random.randn(2, 2)
W_2 = np.random.randn(1, 2)
b_1 = np.zeros((2, 1))
b_2 = np.zeros((1, 1))
```

```python
def sigmoid(A):
    return 1 / (1 + np.exp(-A))
```

```python
def loss (y_true, y_pred):
    return np.mean(np.square(y_true - y_pred))
```

```python
eta = 0.5
epoch = 1000000
for i in range(epoch):
  z_1 = np.dot(W_1, x) + b_1
  a_1 = sigmoid(z_1)
  z_2 = np.dot(W_2, a_1) + b_2
  a_2 = sigmoid(z_2)

  delta_2 = (a_2 - y_true) * (a_2 * (1 - a_2))
  delta_1 = (np.dot(W_2.T, delta_2)) * (a_1 * (1 - a_1))

  grad_W_2 = 0.25*(np.dot(delta_2, a_1.T))
  grad_b_2 = 0.25*np.sum(delta_2, axis=1, keepdims=True)
  grad_W_1 = 0.25*(np.dot(delta_1, x.T))
  grad_b_1 = 0.25*np.sum(delta_1, axis=1, keepdims=True)

  W_2 = W_2 - eta*grad_W_2
  b_2 = b_2 - eta*grad_b_2
  W_1 = W_1 - eta*grad_W_1
  b_1 = b_1 - eta*grad_b_1

y_pred = sigmoid(np.dot(W_2, sigmoid(np.dot(W_1, x) + b_1)) + b_2)
print(y_pred)
print(loss(y_true, y_pred))
```
