```python
import numpy as np
```

```python
class LinearSVM:

  def __init__(self,learning_rate,lambda_param,n_iters):
    self.learning_rate = learning_rate
    self.lambda_param = lambda_param
    self.n_iters = n_iters

  def fit(self,X,y):
    self.w = np.zeros((X.shape[1],1))
    self.b = 0
    y[y==0]=-1

    for epoch in range(self.n_iters):
      for i in range(len(X)):
        if(y[i]*(np.matmul(X[i],self.w)-self.b).item()>=1):
          dJ_dw = 2*self.lambda_param*self.w
          self.w-=self.learning_rate*dJ_dw
        else:
          dJ_dw = 2*self.lambda_param*self.w + ((-y[i]*X[i]).reshape(-1,1))
          dJ_db = y[i].item()
          self.w-=self.learning_rate*dJ_dw
          self.b-=self.learning_rate*dJ_db

  def predict(self,X):
    z = np.matmul(X,self.w)-self.b
    z[z>0]=1
    z[z<0]=-1
    return z
```

# Testing


```python
X = np.array([
    [1, 2],
    [2, 3],
    [2, 0],
    [3, 1]
], dtype=float)

y = np.array([0, 0, 1, 1], dtype=int)
```

```python
svm1 = LinearSVM(0.05,0.3,10)
```

```python
svm1.fit(X,y)
```

```python
svm1.predict(X)
```
