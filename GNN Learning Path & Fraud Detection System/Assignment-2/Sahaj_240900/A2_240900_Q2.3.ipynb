```python
import numpy as np
```

```python
class LinearSVM():
    def __init__(self,learning_rate,n_iters,lambda_param):
        self.learning_rate=learning_rate
        self.n_iters=n_iters
        self.lambda_param=lambda_param
    
    def fit(self,X,y):
        self.w=np.zeros(X.shape[1])
        self.b=0

        self.y=np.where(y==0,-1,1).astype(float)

        for _ in range(self.n_iters): 
            for idx, x_i in enumerate(X):
                if self.y[idx]*(np.dot(x_i,self.w)-self.b)>=1:
                    self.w=self.w-self.learning_rate*(2*self.lambda_param*self.w)
                else:
                    self.w=self.w -self.learning_rate*(2*self.lambda_param*self.w-(1/len(self.y))*self.y[idx]*x_i)
                    self.b=self.b-(1/len(self.y))*self.y[idx]*self.learning_rate

    def predict(self,X):
        y_pred=np.dot(X,self.w)-self.b
        return np.sign(y_pred)
```
