```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/housing_prices - housing_prices.csv')

x_vals = df.iloc[:, 0].values
y_vals = df.iloc[:, 1].values

A = np.c_[np.ones(x_vals.shape[0]), x_vals]

ATA = A.T @ A
ATA_inv = np.linalg.inv(ATA)
ATy = A.T @ y_vals

params = ATA_inv @ ATy
intercept, slope = params[0], params[1]

target_sqft = 2500
estimate = slope * target_sqft + intercept
print(f"OLS Predicted price for {target_sqft} sq ft: ${estimate:.2f}")

plt.figure(figsize=(10, 6))
plt.scatter(x_vals, y_vals, color='blue')

x_plot = np.linspace(x_vals.min(), x_vals.max(), 100)
y_plot = slope * x_plot + intercept

plt.plot(x_plot, y_plot, color='red', label=f'OLS Line: y={slope:.2f}x + {intercept:.2f}')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Housing Prices: OLS Method')
plt.legend()
plt.grid(True)
plt.show()
```
