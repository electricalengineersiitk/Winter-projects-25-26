```python
import numpy as np
import math

X = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

y = [0,1,1,0]

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def dsigmoid(x):
    return x * (1 - x)

w1 = [[0.5, -0.5],
      [0.5, -0.5]]

b1 = [0.0, 0.0]

w2 = [0.5, 0.5]
b2 = 0.0

lr = 0.1

for epoch in range(10000):

    for i in range(4):

        x1 = X[i][0]
        x2 = X[i][1]

        h1 = sigmoid(x1*w1[0][0] + x2*w1[1][0] + b1[0])
        h2 = sigmoid(x1*w1[0][1] + x2*w1[1][1] + b1[1])

        out = sigmoid(h1*w2[0] + h2*w2[1] + b2)

        error = y[i] - out

        d_out = error * dsigmoid(out)

        d_h1 = d_out * w2[0] * dsigmoid(h1)
        d_h2 = d_out * w2[1] * dsigmoid(h2)

        w2[0] = w2[0] + lr * d_out * h1
        w2[1] = w2[1] + lr * d_out * h2
        b2 = b2 + lr * d_out

        w1[0][0] = w1[0][0] + lr * d_h1 * x1
        w1[1][0] = w1[1][0] + lr * d_h1 * x2
        b1[0] = b1[0] + lr * d_h1

        w1[0][1] = w1[0][1] + lr * d_h2 * x1
        w1[1][1] = w1[1][1] + lr * d_h2 * x2
        b1[1] = b1[1] + lr * d_h2

for i in range(4):
    x1 = X[i][0]
    x2 = X[i][1]

    h1 = sigmoid(x1*w1[0][0] + x2*w1[1][0] + b1[0])
    h2 = sigmoid(x1*w1[0][1] + x2*w1[1][1] + b1[1])

    out = sigmoid(h1*w2[0] + h2*w2[1] + b2)

    print(X[i], round(out))
```
