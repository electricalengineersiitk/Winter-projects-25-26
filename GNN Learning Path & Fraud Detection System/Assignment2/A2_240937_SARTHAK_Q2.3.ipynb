```python
import numpy as np

class LinearSVM:

    def __init__(self, lr=0.001, lam=0.01, it=1000):
        self.lr = lr
        self.lam = lam
        self.it = it
        self.w = None
        self.b = 0

    def fit(self, X, y):

        n = len(X)
        m = len(X[0])

        yy = []
        for i in range(len(y)):
            if y[i] == 0:
                yy.append(-1)
            else:
                yy.append(1)

        self.w = np.zeros(m)
        self.b = 0

        for k in range(self.it):
            for i in range(n):

                xi = X[i]
                yi = yy[i]

                s = 0
                for j in range(m):
                    s = s + self.w[j] * xi[j]

                val = yi * (s - self.b)

                if val >= 1:
                    for j in range(m):
                        self.w[j] = self.w[j] - self.lr * (2 * self.lam * self.w[j])
                else:
                    for j in range(m):
                        self.w[j] = self.w[j] - self.lr * (2 * self.lam * self.w[j] - yi * xi[j])
                    self.b = self.b - self.lr * yi

    def predict(self, X):

        res = []
        for i in range(len(X)):
            s = 0
            for j in range(len(X[0])):
                s = s + self.w[j] * X[i][j]

            if s - self.b >= 0:
                res.append(1)
            else:
                res.append(-1)

        return res


X = np.array([[1,2],[2,3],[3,3],[2,1]])
y = np.array([0,0,1,1])

sv = LinearSVM()
sv.fit(X,y)
p = sv.predict(X)
print(p)
```
