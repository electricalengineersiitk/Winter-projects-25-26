# Exploratory Data Analysis & P300 Pipeline

This notebook demonstrates loading the BCI dataset using `moabb`, epoching the data, performing average re-referencing, and plotting the grand average ERP representing the P300 component.

```python
import sys
import os
sys.path.append(os.path.abspath('../src'))
from preprocess import get_clean_data
import matplotlib.pyplot as plt
```

## 1. Data Loading
We load data for Subject 1 from BNCI2014_009 dataset. This automatically handles download, MNE filtering, Notch, and average referencing.

```python
epochs, X, y = get_clean_data('BNCI2014_009', subj=1)
print(epochs)
```

## 2. ERP Visualization
We can plot the ERPs of Target vs Non-Target to observe the P300 component.

```python
target_evoked = epochs['Target'].average()
nontarget_evoked = epochs['NonTarget'].average()

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
target_evoked.plot(axes=axes[0], spatial_colors=True, titles='Target (P300)')
nontarget_evoked.plot(axes=axes[1], spatial_colors=True, titles='Non-Target')
plt.show()
```
