PANDAS Ex2

```python
import pandas as pd
import numpy as np

df.replace(['?', 'n.a', 'N.A', 'NA'], np.nan, inplace=True)

numeric_cols = ["wheel-base", "length", "horsepower", "average-mileage", "price"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print(df)
```

MATPLOT Ex1

```python
import matplotlib.pyplot as plt

df1 = pd.read_csv("/content/company_sales_data.csv")
profit = df1['total_profit']
months = df1['month_number']
plt.figure(figsize=(10,5))
plt.plot(months, profit, marker='o')
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company Profit Per Month")
plt.grid(True)
plt.show()
```

MATPLOT Ex2

```python
months = df1['month_number']
face_cream = df1['facecream']
face_wash = df1['facewash']
bar_width = 0.4
x1 = months
x2 = [m + bar_width for m in months]
plt.figure(figsize=(10,5))

plt.bar(x1, face_cream, width=bar_width, label='Face Cream Sales')
plt.bar(x2, face_wash, width=bar_width, label='Face Wash Sales')
plt.xlabel("Month")
plt.ylabel("Units")
plt.title("Sales of Face Cream & Face Wash")
plt.xticks(months + bar_width / 2, months)
plt.legend()
plt.show()
```
