```python
import numpy as x1
import mne as x2

from moabb.datasets import BNCI2014_009 as x3
from moabb.paradigms import P300 as x4

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as x5
from sklearn.model_selection import cross_val_score as x6
from sklearn.metrics import confusion_matrix as x7
```

```python

print("getting dataset...")
x8=x3()
x9=x4()
y1,y2,y3=x9.get_data(dataset=x8, subjects=[1])

print("data shape:",y1.shape)

def y4(y5):
    y5 = x2.filter.filter_data(y5, sfreq=256, l_freq=0.1, h_freq=30)
    y5 = x2.filter.notch_filter(y5, sfreq=256, freqs=50)
    return y5

y1=y4(y1)

def y6(y7):
    y7=y7[:, :, ::8]
    y7=y7.reshape(y7.shape[0], -1)
    return y7
y8 =y6(y1)
z1= x5()

print("Training...")
z2 = x6(z1, y8, y2, cv=5)

z3 = x1.mean(z2)

print("accuracy:", z3)

z1.fit(y8, y2)

z4=z1.predict(y8)
z5=x7(y2, z4)
print("Confusion Matrix:\n", z5)
def z6(z7, z8, z9):
    if z8 <= 0 or z8 >= 1:
        return 0
    return (x1.log2(z7) + z8*x1.log2(z8) + (1-z8)*x1.log2((1-z8)/(z7-1))) * 60 / z9

a1=36
a2=z3
a3=2.5
a4 = z6(a1, a2, a3)
print("ITR (bits/min):", a4)
print("\n===== FINAL RESULTS =====")
print(f"Accuracy: {z3:.4f}")
print(f"ITR: {a4:.2f} bits/min")
```
